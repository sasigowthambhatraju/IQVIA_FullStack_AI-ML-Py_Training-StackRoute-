```python
# As U can see from above code there are missing values in above ds
# So for childern there are 4 missing values here we'll be replacing with mean or median
# for agent the dtype is float so we can use either mean or median
# but for country it's dtype is object so we need to use mode only in order to replace the null values in the column
# As the company column is having so much nan values out of overall data we can remove the colun entirely becoz it's not impacting
```

```python
mean_childern = df['children'].mean()
mean_childern
```

```python
# Now replacing the null values from childern column in the ds
df['children'].replace(np.nan, mean_childern, inplace=True)
```

```python
df.isnull().sum()
```

```python
# For agent i.e of float dtype
median_agent = df['agent'].median()
median_agent
```

```python
df['agent'].replace(np.nan, median_agent, inplace=True)
```

```python
# As U can see from above lines of code the childern's column null values got replaced now it's for country column
# because it's object type we'll be using mode

country_mode = df['country'].mode().values[0]
country_mode
```

```python
df['country'].replace(np.nan, country_mode, inplace=True)
```

```python
# Checking and removing the outliers using boxplot
df.boxplot(column='lead_time')
```

```python
# Function for removing the outliers from 1column
def remove_outliers(col):
    sorted(col)
    q1, q3 = col.quantile([0.25,0.75])
    iqr = q3-q1
    lower_range = q1-(1.5*iqr)
    upper_range = q3+(1.5*iqr)
    return lower_range, upper_range
```

```python
lowledtime, uppleadtime = remove_outliers(df['lead_time'])
```

```python
df['lead_time'] = np.where(df['lead_time']>uppleadtime, uppleadtime, df['lead_time'])
```

```python
df['lead_time'] = np.where(df['lead_time']<lowledtime, lowledtime, df['lead_time'])
```
